Abstract class

```python
from abc import ABC, abstractmethod

class bank(ABC):
    def loan(self):
        print("Loan Approved ")

    @abstractmethod
    def p_d(self):
        pass

class customer(bank):
    def p_d(self):
        print("Customer details")

c = customer()
c.p_d()
c.loan()
```
