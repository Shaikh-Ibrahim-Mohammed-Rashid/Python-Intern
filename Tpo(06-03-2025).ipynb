```python

x = 10
y = x

print(id(x))
print(id(y))  # x and y point to the same memory location initially.

x = 20  # Changing x does not affect y because integers are immutable.

print(id(x))  # x now points to a different memory location.
print(id(y))  # y still points to the original location (value 10).

# Strings are immutable.
s1 = "hello"
s2 = s1

print(id(s1))
print(id(s2))

s1 = "world" # changing s1 does not affect s2

print(id(s1))
print(id(s2))

# Tuples are immutable
tuple1 = (1, 2, 3)
tuple2 = tuple1

print(id(tuple1))
print(id(tuple2))

# Attempting to modify a tuple will raise an error
# tuple1[0] = 4 # This line will cause a TypeError
```

```python

my_string = "Hello, World!"

# Basic slicing
print(my_string[0:5])  # Output: Hello
print(my_string[7:])   # Output: World!
print(my_string[:5])   # Output: Hello
print(my_string[-6:])  # Output: World!

# Slicing with steps
print(my_string[::2])   # Output: Hlo ol! (every second character)
print(my_string[::-1])  # Output: !dlroW ,olleH (reversed string)

# Slicing with negative indices and steps
print(my_string[-1:-7:-1]) # Output: !dlroW


my_list = [10, 20, 30, 40, 50]

print(my_list[1:4])   # Output: [20, 30, 40]
print(my_list[::-1]) # Output: [50, 40, 30, 20, 10] (reversed list)

# Slicing tuples
my_tuple = (1, 2, 3, 4, 5)
print(my_tuple[2:4]) # Output: (3,4)

#Example of slicing an array
import numpy as np
my_array = np.array([1,2,3,4,5])
print(my_array[1:4]) # Output: [2 3 4]
```

```python
my_string = "hello"
print('h' in my_string)  # Output: True
print('z' in my_string)  # Output: False

my_list = [1, 2, 3, 4, 5]
print(3 in my_list)  # Output: True
print(6 in my_list)  # Output: False

my_tuple = (10, 20, 30)
print(20 in my_tuple)  # Output: True
print(40 in my_tuple)  # Output: False


my_set = {100, 200, 300}
print(200 in my_set) # Output: True
print(400 in my_set) # Output: False

my_dict = {"a": 1, "b": 2, "c": 3}
print("a" in my_dict)  # Checks keys (Output: True)
print(1 in my_dict)   # Output: False
print(1 in my_dict.values()) # Output: True


# 'not in' operator: Returns True if a value is NOT found in the sequence, False otherwise.

print('h' not in my_string)  # Output: False
print('z' not in my_string)  # Output: True

print(3 not in my_list)  # Output: False
print(6 not in my_list)  # Output: True

# Membership check in NumPy arrays

import numpy as np
my_array = np.array([1,2,3,4,5])
print(2 in my_array) #Output: True
print (6 in my_array) # Output: False
```

```python
def analyze_string(input_string):
    """Analyzes a string using various string methods.

    Args:
      input_string: The string to analyze.

    Returns:
      A dictionary containing the results of the analysis.
    """

    results = {}
    results['isalnum'] = input_string.isalnum()
    results['isalpha'] = input_string.isalpha()
    results['isdigit'] = input_string.isdigit()
    results['islower'] = input_string.islower()
    results['isupper'] = input_string.isupper()
    results['istitle'] = input_string.istitle()
    results['isspace'] = input_string.isspace()
    results['startswith_h'] = input_string.startswith('h')
    results['endswith_d'] = input_string.endswith('d')
    return results


# Example usage:
my_string = "hello d"
analysis_results = analyze_string(my_string)
print(analysis_results)


my_string = "hello world"
print(my_string.title())  # Output: Hello World
print(my_string.capitalize()) # Output: Hello world
```
