```python
try:
    # Code that may raise an exception
    a = 10 / 0  # Division by zero error
except ZeroDivisionError:
    print("Error: Division by zero is not allowed!")
```

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Error: Invalid input, please enter a number!")
except ZeroDivisionError:
    print("Error: Division by zero is not allowed!")
```

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ZeroDivisionError:
    print("Cannot divide by zero!")
else:
    print("Result:", result)
```

```python
try:
    file = open("test.txt", "r")
    content = file.read()
except FileNotFoundError:
    print("File not found!")
finally:
    print("Closing file...")
```

```python
def check_age(age):
    if age < 18:
        raise ValueError("You must be 18 or older.")
    return "Access granted!"

try:
    print(check_age(15))
except ValueError as e:
    print("Exception:", e)
```

```python
try:
    lst = [1, 2, 3]
    print(lst[5])  # Index out of range
except IndexError:
    print("Error: Index out of range!")
```

```python
try:
    data = {"name": "Alice", "age": 25}
    print(data["city"])  # Key doesn't exist
except KeyError:
    print("Error: Key not found in dictionary!")
```

```python
import logging

try:
    num = int(input("Enter a number: "))
    result = 100 / num
except Exception as e:
    logging.error(f"An error occurred: {e}")
```

```python
def withdraw(amount):
    if amount > 10000:
        raise ValueError("Withdrawal limit exceeded!")
    return f"Withdrew {amount} successfully"

try:
    print(withdraw(15000))
except ValueError as e:
    print("Exception:", e)
```

```python
try:
    num = int(input("Enter a number: "))
    try:
        result = 10 / num
        print(result)
    except ZeroDivisionError:
        print("Inner Exception: Cannot divide by zero!")
except ValueError:
    print("Outer Exception: Invalid input! Enter a number.")
```

```python
try:
    age = int(input("Enter your age: "))
    assert age >= 18, "You must be 18 or older!"
    print("Access granted.")
except AssertionError as e:
    print("Error:", e)
```
