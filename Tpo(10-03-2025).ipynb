Variable length argument function

```python
def add(*a):
  s = 0
  for i in a:
    s += i
  print(s)
add(1,2,3,4,5)
```

Arbitrary keyword argument

```python
def display(**kwargs):
    for i,j in kwargs.items():
        print(i,j)
display(a=10,b=20,c=30)
```

Passing list as a argument

```python
def print_(food):
  #print(food)
  for i in food :
      print(i, end=" ")

Food = ["Meggie", "pizza"]
print_(Food)
```

Passing dictionary as argument

```python
def print_(food):
  #print(food)
  for i,j in food.items():
      print(i, j, end=" ")

Food = {"Meggie" : "pizza"}
print(type(Food))
print_(Food)
```

Recursion function

```python
def f(n):
    if n == 0:
        return 1
    else:
        return n * f(n-1)
print(f(5))
```

```python
def f(n):
    if n <= 1:
        return n
    else:
        return f(n-1) + f(n-2)

n = 10
print("Fibonachi Series Is : ", end="")
for i in range(n):
    print(f(i),end=" ")
```

```python
def find_smallest_missing_positive(nums):
    n = len(nums)

    num_set = set(nums)

    for i in range(1, n + 2):
        if i not in num_set:
            return i

    return n+1
print(find_smallest_missing_positive([1,2,4]))
```
